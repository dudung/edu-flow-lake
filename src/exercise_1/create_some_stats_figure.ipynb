```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# define input file name and path
rr_path = '../../data/analyzed/admission'
rr_file = rr_path + '/rejection_rate.xlsx'

# read data from file as dataframe
df = pd.read_excel(rr_file)

# create list from dataframe
x1 = list(df['semester'])
x2 = list(df['batches'])
x3 = list(df['applied'])
x4 = list(df['accepted'])
x5 = list(df['rejection'])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x1, x5, 'ro--')
plt.xticks(rotation=90, ha='right')
plt.grid()
plt.ylabel('rejection rate')
plt.xlabel('semester')
plt.ylim([0,1])
plt.savefig(rr_path + '/rejection_rate_semester.png')
```

```python
plt.figure(figsize=(6,4))
plt.plot(x2, x5, 'o')
plt.grid()
plt.ylabel('rejection rate')
plt.xlabel('number of batches')
plt.ylim([0,1])
plt.xticks(range(1,5))
plt.savefig(rr_path + '/rejection_rate_number_of_batches.png')
```

```python
plt.figure(figsize=(6,4))
plt.plot(x1, x2, 'o')
plt.xticks(rotation=90, ha='right')
plt.grid()
plt.ylabel('number of batches')
plt.xlabel('semester')
plt.savefig(rr_path + '/number_of_batches_semester.png')
```

```python
plt.figure(figsize=(6,4))
plt.plot(x1, x2, 'go')
plt.xticks(rotation=90, ha='right')
plt.grid()
plt.ylabel('number of batches')
plt.xlabel('semester')
plt.savefig(rr_path + '/number_of_batches_semester.png')
```

```python
color_map = plt.get_cmap('bwr')
sizes = [i*200 for i in x5]
lws = [i*2 for i in x5]
colors = [color_map(i) for i in x5]

plt.figure(figsize=(6,4))
plt.scatter(x3, x2, marker='o', s=sizes, facecolor=colors, edgecolor='white', lw = lws)
plt.xticks(rotation=90, ha='right')
plt.grid()
plt.ylabel('number of batches')
plt.xlabel('total applicants')
plt.savefig(rr_path + '/number_of_batches_applied_rejection.png')
```
